```python
def crop_image(img, show=True):
    # Binarize the image
    bin_pixels = cv2.threshold(img, 20, 255, cv2.THRESH_BINARY)[1]
   
    # Make contours around the binarized image, keep only the largest contour
    contours, _ = cv2.findContours(bin_pixels, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    contour = max(contours, key=cv2.contourArea)

    # Create a mask from the largest contour
    mask = np.zeros(img.shape, np.uint8)
    cv2.drawContours(mask, [contour], -1, 255, cv2.FILLED)
   
    # Use bitwise_and to get masked part of the original image
    out = cv2.bitwise_and(img, mask)
    
    # get bounding box of contour
    y1, y2 = np.min(contour[:, :, 1]), np.max(contour[:, :, 1])
    x1, x2 = np.min(contour[:, :, 0]), np.max(contour[:, :, 0])
    
    x1 = int(0.99 * x1)
    x2 = int(1.01 * x2)
    y1 = int(0.99 * y1)
    y2 = int(1.01 * y2)
    
    if show:
        plt.imshow(out[y1:y2, x1:x2], cmap="gray") ; 

    return out[y1:y2, x1:x2]
```

```python
def process(f, size=512, extension="png"):
    patient = f.split('\\')[-2]
    image = f.split('\\')[-1].split('.')[0]

    dicom = pydicom.dcmread(f)
    img = dicom.pixel_array

    img = (img - img.min()) / (img.max() - img.min())
    img *= 255
    img = np.uint8(img)
    
    if dicom.PhotometricInterpretation == "MONOCHROME1":
        img = 1 - img

    img = cv2.resize(img, (size, size))
    img = crop_image(img, show=False)
    img = cv2.resize(img, (640, 1280))
    # plt.figure(figsize=(5, 5))
    # plt.imshow(img, cmap="gray")
    # plt.title(f"original image for {patient} {image}")
    # plt.show()
    # print(f"./{patient}_{image}.png")
    cv2.imwrite( f"./img/{patient}_{image}.png", img)
```

```python
slM=defaultdict(list)
srM=defaultdict(list)
slC=defaultdict(list)
srC=defaultdict(list)
slc = defaultdict(int)
src = defaultdict(int)
df = pd.read_csv("./breast-level_annotations.csv")
```

```python
from PIL import Image
```

```python
for i,row in tqdm(df.iterrows()):
    # print(row["study_id"])
    if row["laterality"] == "L":
        if row["view_position"]=="CC":
            slC[row["study_id"]].append(row["image_id"])
        else:
            slM[row["study_id"]].append(row["image_id"])
        if row["breast_birads"] == "BI-RADS 5":
            slc[row["study_id"]]=1
        else:
            slc[row["study_id"]]=0
    else:
        if row["view_position"]=="CC":
            srC[row["study_id"]].append(row["image_id"])
        else:
            srM[row["study_id"]].append(row["image_id"])
        if row["breast_birads"] == "BI-RADS 5":
            src[row["study_id"]]=1
        else:
            src[row["study_id"]]=0
        
```

```python
def join(png1,png2,path,name,flag='horizontal'):
    """
    :param png1: path
    :param png2: path
    :param flag: horizontal or vertical
    :return:
    """
    img1, img2 = Image.open(png1), Image.open(png2)
    i1 = np.array(img1)
    i2 = np.array(img2)
    if png1 == png2:
        i2 = np.zeros_like(i2)
#     size1, size2 = img1.size, img2.size
    if flag == 'horizontal':
        im = np.concatenate((i1,i2),axis=1)
        im = Image.fromarray(im)
        im.save(f'{path}/{name}')
    elif flag == 'vertical':
        im = np.concatenate((i1,i2),axis=0)
        im = Image.fromarray(im)
        im.save(f'./{path}/{name}')
```

```python
for i in tqdm(slc):
    # print(i)
    CC = slC[i]
    MLO = slM[i]
    # print(len(CC))
    for c in CC:
        for m in MLO:
            if slc[i]==1:
                join(f"./img/{i}_{c}.png",f"./img/{i}_{m}.png",f"./cancer",f"{i}_{c}.png")
            else:
                join(f"./img/{i}_{c}.png",f"./img/{i}_{m}.png",f"./normal",f"{i}_{c}.png")
    
```
